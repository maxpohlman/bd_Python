```python


import pandas as pd
#might need to install
#conda install pandas


survey = pd.read_csv('surveys.csv')
sp_counts = survey.groupby('species').record_id.count()
%matplotlib inline

#plot num of each species
sp_counts.plot(kind='bar')
```

```python
#num of species in each plot
sp_plot = survey.groupby('plot').species.nunique()
sp_plot.plot(kind='bar')
```

```python
sp_plot = survey.groupby('plot').wgt.mean()
sp_plot.plot(kind='bar')
```

```python
sp_plot = survey.groupby('plot')['wgt'].mean()
sp_plot.plot(kind='bar')
```

```python
#number of individual species per plot
sp_plot = survey.groupby(['species','plot'],as_index=False).record_id.count()

sp_plot_wide = sp_plot.pivot(index = 'plot', columns = 'species', values = 'record_id')
```

```python

my_plot = sp_plot_wide.plot(kind = 'bar', stacked = True, legend = False, 
                  title = "Species counts per plot")
my_plot.set_xlabel("Plot")
my_plot.set_ylabel("Counts by species")
```

```python
spp = pd.read_csv('species.csv',keep_default_na=False, na_values=[""])
```

```python
survey = pd.read_csv('surveys.csv',keep_default_na=False, na_values=[""])
```

```python
#join datasets given app - species_id and survey - species
#left join - keep all survey data - add species info if known
merged_left = pd.merge(left = survey, right = spp, left_on = 'species', right_on = 'species_id', how = 'left')
```

```python
survey.head()
```

```python
#innerjoin - same as left but don't include NaNs
merged_inner = pd.merge(left = survey, right = spp, left_on = 'species', right_on = 'species_id')
```

```python

merged_outer = pd.merge(left = survey, right = spp, left_on = 'species', right_on = 'species_id', how = 'outer')
```

```python
merged_outer.head()
```

```python
sp_plot = merged_inner.groupby('plot')['taxa'].count()
sp_plot.plot(kind='bar')
```

```python
!curl -O https://raw.githubusercontent.com/mperignon/python-lessons/gh-pages/data/plots.csv
pdata = pd.read_csv('plots.csv',keep_default_na=False, na_values=[""])
```

```python
pdata.head()
```

```python
pmerge = pd.merge(left = merged_inner, right = pdata, left_on = 'plot', right_on = 'plot_id', how = 'left')
```

```python
pmerge.head()
```

```python
cdata = pmerge[pmerge.plot_type == 'Control']
rdata = pmerge[pmerge.plot_type == 'Rodent Exclosure']
#cp_plott = cdata.groupby(('plot','taxa']).count()
#cp_plott.plot(kind='bar')

sp_plot = cdata.groupby(['taxa','plot'],as_index=False).record_id.count()
sp_plot_wide = sp_plot.pivot(index = 'plot', columns = 'taxa', values = 'record_id')
my_plot = sp_plot_wide.plot(kind = 'bar', stacked = True, legend = True, 
                  title = "Species counts per plot")
my_plot.set_xlabel("Plot")
my_plot.set_ylabel("Counts by taxa")
```

```python
sp_plot = rdata.groupby(['taxa','plot'],as_index=False).record_id.count()
sp_plot_wide = sp_plot.pivot(index = 'plot', columns = 'taxa', values = 'record_id')
my_plot = sp_plot_wide.plot(kind = 'bar', stacked = True, legend = True, 
                  title = "Species counts per plot")
my_plot.set_xlabel("Plot")
my_plot.set_ylabel("Counts by taxa")
```
